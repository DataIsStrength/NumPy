NumPy-数组的切片索引和形状修改

```python
#导入NumPy包
import numpy as np
```

```python
#案例：随机生成8只股票2周的交易日涨幅数据

#生成形状为(8,10)的标准正态随机数组
stock_change=np.random.normal(loc=0,scale=1,size=(8,10))
print(stock_change)
```

```python
#获取第一个股票的前3个交易日的涨跌幅数据
#切片第1列前3行
print(stock_change[0,0:3])
```

```python
#三维数组索引
a1=np.array([[[1,2,3],[4,5,6]],[[8,9,10],[11,12,13]]])
print(a1,'\n',a1.shape)
#索引元素10（第2个二维数组，第1行，第3列）并修改
a1[1,0,2]=0
print(a1)
```

```python
#.reshape()（不改变元素顺序,返回改变后的数组，但不改变原来的数组）
stock_change.reshape((10, 8))
```

```python
#.resize(不改变元素顺序,无返回值，改变原来的数组)
stock_change.resize((10, 8))
print(stock_change)
```

```python
#转置
stock_change.T
```

```python
#一维化
stock_change.flatten()
```
