NumPy-统计运算

```python
#导入NumPy包
import numpy as np
```

```python
#案例：随机生成8只股票2周的交易日涨幅数据

#生成形状为(8,10)的标准正态随机数组
stock_change=np.random.normal(loc=0,scale=1,size=(8,10))
print(stock_change)
```

```python
#前四只股票前四天的最大涨幅

array=stock_change[:4,:4]
print(array)

#.max()
print(array.max())

#np.max()
print(np.max(array))
```

```python
#行、列最值

#前4只股票第1天的最大涨幅
#列最大值（轴参数设置为0）
print(array.max(axis=0))

#前4天第1只股票的最大涨幅
#行最大值（轴参数设置为1）
print(np.max(array,axis=1))
```

```python
#最值索引

#前4天第1只股票的最大涨幅的索引值（从0开始计算）
np.argmax(array, axis=1)
```
