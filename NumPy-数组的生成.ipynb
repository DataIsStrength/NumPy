NumPy-数组的生成

```python
#导入NumPy包
import numpy as np
```

```python
#生成n维数组
score=np.array([[80,89,86,67,79],
                [78,97,89,67,81],
                [90,94,78,67,74],
                [91,91,90,67,69],
                [76,87,75,67,86],
                [70,79,84,67,84],
                [94,92,93,67,64],
                [86,85,83,67,80]])
```

```python
#从现有数组生成数组

#深复制（申请新的内存地址）
#np.array()
data1=np.array(score)
#np.copy()
data2=np.copy(score)

#浅复制（增加一个指向既存内存地址的指针）
data3=np.asarray(score)

#对比修改原数组之后的变化
score[0,0]=0
print(score)
print(data1)
print(data2)
print(data3)
```

```python
#生成0和1的数组

#生成0数组
print(np.zeros(shape=(3,4),dtype='float32'))
#生成1数组
print(np.ones(shape=(3,4),dtype='float32'))
```

```python
#生成等差数组

#[0,10]生成6个元素的等差数组
print(np.linspace(0, 10, 6))
#[0,11)生成步长为2的等差数组
print(np.arange(0, 11, 2))
```

```python
#生成均匀分布的随机数组

#导入Matplotlib包
import matplotlib.pyplot as plt

#生成[-1,1]的均匀分布数组，样本容量为1000000
data1=np.random.uniform(low=-1,high=1,size=1000000)
#可视化
plt.figure(figsize=(20, 8),dpi=80)
#绘制组数为1000的直方图
plt.hist(data1,1000)
plt.show()
```

```python
#生成正态分布的随机数组

#生成标准正态分布数组（loc参数表示均值，scale参数表示标准差），样本容量为1000000
data2=np.random.normal(loc=0,scale=1,size=1000000)
#可视化
plt.figure(figsize=(20,8),dpi=80)
#绘制组数为1000的直方图
plt.hist(data2,1000)
plt.show()
```
